Equação do pêndulo simples de comprimento $l$ e ângulo $\theta$ com a vertical:
$$ \frac{d^2 \theta}{dt^2} = - \frac{g}{l} \sin \theta $$


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
θ = np.linspace(0,20,100+1)
θ_rad = np.deg2rad(θ)
lin, = plt.plot(θ,θ_rad)
nlin, = plt.plot(θ,np.sin(θ_rad))
```
